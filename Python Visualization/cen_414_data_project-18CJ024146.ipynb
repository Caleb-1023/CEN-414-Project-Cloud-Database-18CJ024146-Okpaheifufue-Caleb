```python
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
from statistics import mean, variance, stdev, mode, median
import seaborn as sns
from matplotlib import style
sns.set()


data_set = pd.read_csv('UNdata_Export.csv')
data_set=data_set.set_index('AFRICAN NATIONS')
data_set
```

```python
data_set
```

```python
#Bar chart plot for each nation per year
for d_year in range(2010, 2019):
    data_set.plot.bar(y=str(d_year), figsize=(11,11), title=str(d_year))
    plt.figure(figsize=(50,50))
```

```python
years = ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
```

```python
#Mean share per year
data_set[years].mean()

x = years
y = data_set[years].mean()

plt.bar(x,y)
plt.show()

data_set[years].mean()
```

```python
#Standard Deviation of share per year
data_set[years].std()

x = years
y = data_set[years].std()

plt.bar(x,y)
plt.show()

data_set[years].std()
```

```python
#Variance of share per year
data_set[years].var()
```

```python
#Mode of share per year
data_set[years].mode
```

```python
#Stacked bar chart 
data_set.groupby('AFRICAN NATIONS').median().plot.bar(stacked=True)

plt.title('Median')
```

```python
#Median of share per year
data_set[years].median
```

```python
iqrs =[]
for year in range(2010,2019):
    iqrs.append(data_set[str(year)].quantile(0.75) - data_set[str(year)].quantile(0.25))
    
    
    
plt.ylabel("Interquatile range")
plt.xlabel("Years")
plt.title("INTERQUATILE RANGE FIXED SUBSCRIPTION")
plt.plot(range(2010,2019),iqrs)
```
